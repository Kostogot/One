# Предобработка

```python
answers = {} # Let's create a dictionary for answers
data['profit'] = data.revenue-data.budget # create new column
 
# calculate count and profit films by genres 
def genres():
  genres_counts = {} # dictionary for calculate genres count
  genres_profit = {} # and calculate profit count
  for index in data.index:
    genres = data.genres[index] 
    for genre in genres.split('|'):
      genres_counts.setdefault(genre,0) # create new key
      genres_counts[genre] += 1         # count
      genres_profit.setdefault(genre,0) # create new key
      if data.profit[index] > 0:
        genres_profit[genre] += 1       # count
  
  genres_data = pd.DataFrame({'genre':list(genres_counts.keys()),\
                              'counts':list(genres_counts.values()),\
                              'profit':list(genres_profit.values())})
  return(genres_data)
 
# calculate films in month
def films_in_month():
  films = 0
  films_month = {}
  for month in range(1,13):
    for date in data.release_date:
      if month==int(date[:date.find('/')]):
        films += 1
    films_month.setdefault(month, films) # create new key
    films = 0
  films_month = pd.Series(films_month) 
  return(films_month)
 
# calculate mean of original title or overwiew films
def companies_base(title=0):
  company_list = {}
  for index in data.index:
    companies = data.production_companies[index] 
    for company in companies.split('|'): 
      company_list.setdefault(company,[]) # new key
      values = company_list[company] # old value
      if title>0: # add new value
        values.append(len(data.original_title[index]))
      else:
        values.append(len(data.overview[index]))
      company_list[company] = values
  
  #calculate mean 
  for company in company_list:
    sum = 0
    for value in company_list[company]:
      sum += value
    company_list[company] = sum/len(company_list[company])
 
  company_data = pd.DataFrame({'company':list(company_list.keys()),\
                               'value':list(company_list.values())})
  return(company_data)
```

# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers[27] = '27. Daniel Radcliffe & Rupert Grint'
```

```python
# create dictionary of actors and films
casts_list = {}
for index in data.index:
  casts = data.cast[index].split('|')
  for actor in casts:
    casts_list.setdefault(actor,[])  # new key
    films = casts_list[actor]  # old value
    films.append(data.original_title[index])  # add new value
    casts_list[actor] = films  # save
 
cast_data = {} 
length = len(casts_list)
films = list(casts_list.values())
casts = list(casts_list.keys())
 
# cycle by two row casts_list and films in it
for index in range(length-1):
  for second_index in range(index+1, length):
    for first_film in films[index]:
      for second_film in films[second_index]:
        if first_film == second_film:
          two_actor = casts[index] + ' & '\
                    + casts[second_index]
          cast_data.setdefault(two_actor,0)  # create new key
          cast_data[two_actor] += 1  # count two actors
 
cast_data = pd.DataFrame({'actors':list(cast_data.keys()),\
                          'counts':list(cast_data.values())}) 
maximal = cast_data.counts.max()
list(cast_data[cast_data.counts==maximal].actors)
```
