```
def game_core (number):
    '''First, we set any random number as one of the boundaries of the interval
        in which the desired number can be found,
        then each time we reduce the interval by half.
       The function takes a envisioned number and returns the number of attempts'''
    count = 1
    predict = np.random.randint (1,101)
    left = 0 # Left border of the search interval
    right_0 = 100 # Right border
    right_1 = right_0 # The previous value of the right border is assigned to the next
 
    if number > predict: # We set one of the boundaries equal to the estimated number
        left = predict
    else:
        right_1 = predict
    
    while number != int(right_1): # Decrease the interval until we guess the number
        if number < right_1:
            right_0 = right_1 # Keep the right border
            right_1 -= (right_1 - left)/2 # Move the border twice to the left
            count += 1
        else:
            left = right_1 # Left border becomes right
            right_1 = right_0 # Right takes the previous value
    return (count) # Exit the loop if you guessed right
 
 
def score_game (game):
    '''Run the game 1000 times to see how quickly the game guesses the number'''
    count_ls = []
    np.random.seed(1) # Fix RANDOM SEED
    random_array = np.random.randint(1,101,size=(1000))
    
    for number in random_array:
        count_ls.append(game(number))
    
    score = int(np.mean(count_ls))
    print(f"The algorithm guesses the number on average for {score} attempts")
    return(score)
 
# Run
import numpy as np
score_game(game_core)
```
